```python
#Zaimportuj moduł z pakietu matplotlib
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
```

# utwórz wykresy funkcji liniowej, kwadratowej, trzeciego rzędu oraz wybranej funkcji trygonometryczne

### Liniowa x

```python
#Naryuj 2 subloty i ustaw przestrzeń pomiędzy nimi 
fig, ax_lst = plt.subplots(2)  # a figure with a 2x2 grid of Axes
fig.subplots_adjust(left=None, bottom=None, right=None, top = 2)

#utworz 2 ciągi liczb gdzie jedna będzie miała wartości od 0 do 2 i pomiedzy nimi będzie 98 licz (iteruj od 0 do 2)
x = np.linspace(0, 2, 100)
y = np.linspace(0, 10, 100)
#na pierwszym sublpocie narysuj w/w ciągi
ax_lst[0].plot(x, y , label='Wynik funkcji')
ax_lst[0].set(xlabel = 'oś x') #ustaw x label czyli to co na dole
ax_lst[0].set(ylabel = 'oś y')  #ustaw x label czyli to co po prawej
ax_lst[0].set_title("Funkcja liniowa ") #ustaw tytuł osi
ax_lst[0].legend(loc="upper right")

#To samo co wyżej tylko narysuj wykres dla funkcji -x * 5
b = 5
x = range(0, 20)  # lista argumentów x
y = []  # lista wartości
for i in x:
    y.append((-1 * i) * 5)
    
ax_lst[1].plot(x, y , label='Wynik funkcji')
ax_lst[1].set_title('Wykres f(x) = -x + 5')
ax_lst[1].grid(True)
ax_lst[1].legend(loc="upper right")
```

### Kwadratowa x^2

```python
#Naryuj 2 subloty i ustaw przestrzeń pomiędzy nimi 
fig, ax_lst = plt.subplots(2)  # a figure with a 2x2 grid of Axes
fig.subplots_adjust(left=None, bottom=None, right=None, top = 2)

#utworz 2 ciągi liczb jeden z  100 elementami pomiędzy  -10 a 10. Drugi który bedzie posiadał wartości podniesione do kwadratu
##tzn y = x^2
x = np.linspace(-10,10,100)
y = x**2
#na podstawie funkcj y=x^2 narysuj wykres 
ax_lst[0].plot(x,y, 'r',label='Wynik funkcji')
ax_lst[0].set(xlabel = 'oś x') #ustaw x label czyli to co na dole
ax_lst[0].set(ylabel = 'oś y')  #ustaw x label czyli to co po prawej
ax_lst[0].set_title("Funkcja Kwadratowa 1") #ustaw tytuł osi
ax_lst[0].legend(loc="upper right")
#Funkcja kwadratowa 2 
x = np.linspace(0, 10, 100)
ax_lst[1].plot(x, x**2,label='Wynik funkcji')
ax_lst[1].set(xlabel = 'oś x') #ustaw x label czyli to co na dole
ax_lst[1].set(ylabel = 'oś y')  #ustaw x label czyli to co po prawej
ax_lst[1].set_title("Funkcja Kwadratowa 2") #ustaw tytuł osi
ax_lst[1].legend(loc="upper right")
```

### Trzeciego rzędu x^3

```python
#Naryuj 2 subloty i ustaw przestrzeń pomiędzy nimi 
fig, ax_lst = plt.subplots(2)  # a figure with a 2x2 grid of Axes
fig.subplots_adjust(left=None, bottom=None, right=None, top = 2)

#utworz 3 ciągi liczb jeden z  100 elementami pomiędzy  -10 a 10. Drugi który bedzie posiadał wartości podniesione do potęgi 3
##tzn y = x^3
x = np.linspace(-10,10,100)
y = x**3
#na podstawie funkcj y=x^3 narysuj wykres 
ax_lst[0].plot(x,y, 'r',label='Wynik funkcji')
ax_lst[0].set(xlabel = 'oś x') #ustaw x label czyli to co na dole
ax_lst[0].set(ylabel = 'oś y')  #ustaw x label czyli to co po prawej
ax_lst[0].set_title("Funkcja trzeciego rzędu 1") #ustaw tytuł osi
ax_lst[0].legend(loc="upper right")
#Funkcja Trzeciego rzędu 2 
x = np.linspace(0, 10, 100)
ax_lst[1].plot(x, x**3, label='Wynik funkcji')
ax_lst[1].set(xlabel = 'oś x') #ustaw x label czyli to co na dole
ax_lst[1].set(ylabel = 'oś y')  #ustaw x label czyli to co po prawej
ax_lst[1].set_title("Funkcja  trzeciego rzędu 2") #ustaw tytuł osi
ax_lst[1].legend(loc="upper right")
```

### Funkcja trygonometryczna

```python
#Naryuj 2 subloty i ustaw przestrzeń pomiędzy nimi 
fig, ax_lst = plt.subplots(2)  # a figure with a 2x2 grid of Axes
fig.subplots_adjust(left=None, bottom=None, right=None, top = 2)

#utworz 2 ciągi jeden z liczbami rozmieszczonymi liniowo z użyciem pi drugi y = sin(x)
x = np.linspace(-np.pi,np.pi,100)
y = np.sin(x)

#na podstawie funkcj y=sin(x) narysuj wykres 
ax_lst[0].plot(x,y, 'r',label='Wynik funkcji')
ax_lst[0].set(xlabel = 'oś x') #ustaw x label czyli to co na dole
ax_lst[0].set(ylabel = 'oś y')  #ustaw x label czyli to co po prawej
ax_lst[0].set_title("Funkcja trygonometryczna 1") #ustaw tytuł osi
ax_lst[0].legend(loc="upper right")

#Utworz liczby od 0 do 100 inkrementowane co 0.4 i oblicz sinus każdego z nich
x = np.arange(0, 100, .4)
y = np.sin(x)
ax_lst[1].plot(x, y,label='Wynik funkcji')
ax_lst[1].set(xlabel = 'oś x') #ustaw x label czyli to co na dole
ax_lst[1].set(ylabel = 'oś y')  #ustaw x label czyli to co po prawej
ax_lst[1].set_title("Funkcja trygonometryczna 2") #ustaw tytuł osi
ax_lst[1].legend(loc="upper right")
```

# utwórz wykres słupkowy (bar chart) na podstawie danych z pliku ‘samochody1tys.csv’ (lab. nr 2). Może to być np. liczebność danej marki samochodu. Można użyć pakietów numpy i pandas, jeśli jest taka potrzeba

```python
# z samym wykorzystaniem pandy
cars = pd.read_csv('samochody1tys.csv') #pobierz dane
df = cars.marka.value_counts().to_frame(); #policz ilości marek a następnie zapisz to do obiektu frame
df.plot.bar(y='marka') # narysuj na podstawie ilości marek 
plt.ylabel('Ilość')
plt.xlabel('Marki')
plt.title('Ilość marek w pliku')
```

```python
carsSorted = cars['marka'].value_counts().keys().tolist()
y_pos = np.arange(len(carsSorted))
valuesSorted = cars['marka'].value_counts().tolist()
plt.bar(y_pos, valuesSorted,label='Ilości wystąpień poszczególnych marek')
plt.ylabel('Ilość')
plt.xlabel('Marki')
plt.title('Ilość marek w pliku')
plt.xticks(y_pos, carsSorted, rotation='vertical')
plt.legend(loc="upper right")
plt.show()
```

# utwórz wykres kołowy (pie chart) na podstawie ww. pliku, ale z innymi danymi niż do wykresu słupkowego

```python
#Utwórz wykres kołowy który pokaże ilość aut wyprodukowanych w danych latach oraz ich procent co do całości
#pobierz wartości roku producji i zapisz do listy
labels  = cars['rok_produkcji'].value_counts().keys().tolist()
#pobierz ilość wystąpień roku producji i zapisz do listy
sizes  = cars['rok_produkcji'].value_counts().tolist()

#narysuj wykres używając ilości wystąpień i obróć go o 90 stopni
patches, texts = plt.pie(sizes, startangle=90)
plt.legend(patches, labels) #utwórz legende na podstawie lat
plt.axis('equal') # zmień położenie wykresu
plt.ylabel('Ilość')
plt.xlabel('Wykres')
plt.title('Ilość roków produkcji w pliku')
plt.show()
```

```python
#pobierz wartości roku producji i zapisz do listy
labels  = cars['rok_produkcji'].value_counts().keys().tolist()
#pobierz ilość wystąpień roku producji i zapisz do listy
sizes  = cars['rok_produkcji'].value_counts().tolist()


#ponieważ wykres nakłada na siebie napisy w przypadku dużej ilości małych wyników usuń 15 ostatnich, zsumuj je 
#i dodaj na koniec wykresu
for x in range(14):
    labels = labels[:-1]
    
sum = 0
for v in range(14):
     sizes = sizes[:-1]
     sum = sum + sizes[-1]
  
labels.append('inne')
sizes.append(sum)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels,shadow=True,autopct='%1.1f%%') #Narysuj wykres i dodaj procenty do każdego roku
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.xlabel('Wykres')
plt.title('Ilość roków produkcji w pliku')
plt.show()
```

# utwórz histogram na podstawie ww. pliku, ale z innymi danymi niż poprzednie

```python
#Narysuj histogram na podstawie przebiegów, a następnie utwórz 30  krawędzi żeby poprawić wizualizację danych
przebiegi  = cars['przebieg'].value_counts().keys().tolist()
plt.hist(przebiegi,30,label='Ilości wystąpień poszczególnych przebiegów')
plt.ylabel('Przebiego')
plt.xlabel('Ilości')
plt.title('Ilość przebiegów z pliku')
plt.legend(loc="upper right")
plt.show()
```
