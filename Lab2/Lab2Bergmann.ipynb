## Scalanie 

```python
#pobierz część wierszy a następnie je scal
auta = [cars[2:4], cars[6:8],cars[3:7]]
pd.concat(auta)
```

```python
#Weź wartości kolumn rok i rodziaj, następnie weź 3 pierwsze rekordy po czym weź rekordy z id od 10:20 poczym dodaj je do siebie
czescpierwszychaut = pd.DataFrame(cars,columns=['rok_produkcji','rodzaj_silnika'])
pierwszeAuta = czescpierwszychaut.iloc[3]
czescpierwszychaut = czescpierwszychaut.iloc[10:20]
czescpierwszychaut.append(pierwszeAuta)
print(czescpierwszychaut)
```

## Zmiana Kształtu

```python
#Dokonaj kompresji Data Frame
#czyli z 
#bar   one     0.029399 -0.542108
#      two     0.282696 -0.087302
#baz   one    -1.575170  1.771208
#      two     0.816482  1.100230
#zrób
#bar    one     A    0.029399
#               B   -0.542108
#       two     A    0.282696
#               B   -0.087302

carsFrame = pd.DataFrame(cars)
print('po wywołaniu stacka')
print(carsFrame)
print('------------------------')
stacked = carsFrame.stack()
print('po wywołaniu unstacka')
unstacked = stacked.unstack()
```

### Pivot table

```python
#Na podstawie marki i ceny wskaż rok produkcji auta
carsTable = pd.DataFrame(cars)
carsPivotTable = pd.pivot_table(carsTable,values='rok_produkcji',index=['cena'],columns=['marka'])
print(carsPivotTable)
```

## Grupowanie 

```python
#Groupuj auta na podstawie marki i ceny a następnie zsumuj
cars.groupby(['marka','cena']).sum()
```

# Kategorie

```python
#Zamień float na typ 'category', a następnie zamiast wyświetlać nazwe silnika napisz opis typu czyli lpg jest najtańasze diesel nadaje się do najdłużej trasy
carsAsDF = pd.DataFrame(cars)
carsAsDF["rodzajeSilnika"] = carsAsDF["rodzaj_silnika"].astype("category")
carsAsDF["rodzajeSilnika"] = carsAsDF["rodzajeSilnika"].cat.set_categories([ "Normlany","Inny","Najdłuższa traca", "Najtańszy","Nowy","Drogi"],rename=True)
carsAsDF["rodzajeSilnika"]
```
